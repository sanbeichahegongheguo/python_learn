```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# 数据的直方图
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
#添加标题
plt.title('Histogram of IQ')
#添加文字
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
```

```python
from ggplot import meat

meat = meat.dropna(thresh=800, axis=1) # 如果某一列的有效数据少于800则删除那一列
ts = meat.set_index(['date'])
ts.head(10)
```

```python
import pandas as pd
from ggplot import *
meat_lng=pd.melt(meat[['date','beef','pork','broilers']],id_vars='date')
ggplot(aes(x='date',y='value',color='variable'),data=meat_lng)+geom_point(color='red')
```

```python
def floor_decade(date_value):
    "Takes a date. Returns the decade."
    return (date_value.year // 10) * 10

import pandas as pd
floor_decade(pd.to_datetime('2013-10-09'))
```

```python
ts.groupby(floor_decade).sum()
```

```python
from ggplot import *
%matplotlib inline
by_decade = ts.groupby(floor_decade).sum()

by_decade.index.name = 'year'

by_decade = by_decade.reset_index()

ggplot(by_decade, aes('year', weight='beef')) + \
    geom_bar() + \
    scale_y_continuous(labels='comma') + \
    ggtitle('Head of Cattle Slaughtered by Decade')
```

```python
by_decade_long = pd.melt(by_decade, id_vars="year")

ggplot(aes(x='year', weight='value', color='variable'), data=by_decade_long) + \
geom_bar() + \
ggtitle("Meat Production by Decade")
```
